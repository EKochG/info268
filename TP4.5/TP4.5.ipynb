<b>1) A partir de los datos de 2017, dibujar un Scatter Plot que muestra la relación entre la variable “Happiness Score” y “Generosity”. 
ver: https://matplotlib.org/gallery/lines_bars_and_markers/scatter_with_legend.html </b>

```python
import pandas as pd
import matplotlib.pyplot as plt
a=['X:Happiness Score', 'Y: Generosity']
hg =pd.read_csv('2017.csv', sep=",")
hpg=hg[['Happiness.Score','Generosity']]
plt.scatter(hpg['Happiness Score'], hpg['Generosity'], c='red', s=60.0, label=a,
               alpha=0.3, edgecolors='none')
plt.legend()
plt.grid(True)
plt.show()
```

<b> 2) En otros Scatter plots, mostrar la relación del “Happiness Score” con las variables “Trust (Government corruption)” y “Economy (GDP per capita)” </b>

```python

plt.scatter(hg['Happiness Score'], hg['Trust..Government.Corruption.'], c='red', s=60.0, label="Relacion happiness/ Trust",
               alpha=0.3, edgecolors='none')

plt.scatter(hg['Happiness Score'], hg['Economy..GDP.per.Capita.'], c='Green', s=60.0, label="Relacion happiness/ economy",
               alpha=0.3, edgecolors='none')

plt.legend()
plt.grid(True)
plt.show()
```

<b>3) Utilizando una función “Group by”, ¿Cuál es puntaje promedio de cada variable por cada región del mundo? Visualizar el resultado con un histograma.
 </b>

```python

hist=hg.groupby('Country').mean()
plt.rcParams['figure.figsize']=(28,20)
hist.plot(kind='bar',stacked=True)
```

<b>4) Utilizando una función “Merge” para agrupar los archivos 2015, 2016 y 2017, ¿Cómo se comporta la variable “Happiness Score” por cada país y por cada región del mundo? Visualizar el resultado con un line chart o un histograma.
 </b>

```python
hg0=pd.read_csv('2015.csv', sep=",")
hg1=pd.read_csv('2016.csv', sep=",")
hg2=pd.read_csv('2017.csv', sep=",")
hg01=pd.merge(hg0[['Country','Region','Happiness Score']],hg1[['Country','Happiness Score']], on='Country')
mf=pd.merge(hg01,hg2[['Country','Happiness Score']], on='Country')
#mf.plot(kind='hist',stacked=True) #no muestra los nombres
mf.groupby('Country').mean().plot(kind='bar',stacked=True)
#Happiness Score_x=2015, Happiness Score_y=2016, Hapiness_score=2017
```

```python
mf.groupby('Region').mean().plot(kind='bar',stacked=True)
```
