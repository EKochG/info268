<b>Preguntas:</b>

<p>1) En un histograma o en un line plot, visualizar el número de tweets conteniendo la
palabra “goal”, cada minuto y cada 10 minutos. (BONUS: gestionar las variantes
“goaaal”, “goooaallll”, etc. utilizando expresiones regulares).</p>

```python
from elasticsearch import Elasticsearch
import pandas as pd
import matplotlib.pyplot as plt
import pprint
import datetime

def timestamp2date(x):
    y = datetime.datetime.fromtimestamp(x/1000).strftime('%H:%M:%S')
    return y

es = Elasticsearch()
res1m = es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "1m" \
                                                                                 }}},"query": {"match": {'text':'goal'}}})
res10m=es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "10m" \
                                                                                 }}},"query": {"match": {'text':'goal'}}})
t1 = pd.DataFrame(res1m['aggregations']['tweets_over_time']['buckets'])
t10=pd.DataFrame(res10m['aggregations']['tweets_over_time']['buckets'])
t1=t1.rename(columns={'key': 'timestamp', 'doc_count': 'nb_tweets'})
t10=t10.rename(columns={'key': 'timestamp', 'doc_count': 'nb_tweets'})
t1['timestamp']=t1['timestamp'].apply(timestamp2date)
t10['timestamp']=t10['timestamp'].apply(timestamp2date)
t1[:30].plot(kind="bar",stacked=True,x="timestamp")
```

```python
t10.plot(kind="bar",stacked=True,x="timestamp")
```

<p>2) En un histograma o en un ​ line plot , ​ visualizar dos series de datos:
a) el número de tweets que hablan de ‘liverpool’ por cada minuto,
b) el número de tweets que hablan de ‘madrid’ por cada minuto.
¿A qué minuto se habla más de Madrid?</p>

```python
resmadrid = es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "1m" \
                                                                                 }}},"query": {"match": {'text':'madrid'}}})
resliverpool= es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "1m" \
                                                                                 }}},"query": {"match": {'text':'liverpool'}}})
dm=pd.DataFrame(resmadrid['aggregations']['tweets_over_time']['buckets'])
lv=pd.DataFrame(resliverpool['aggregations']['tweets_over_time']['buckets'])
dm=dm.rename(columns={'key': 'timestamp', 'doc_count': 'nb_tweets_madrid'})
lv=lv.rename(columns={'key': 'timestamp', 'doc_count': 'nb_tweets_liverpool'})
dm['timestamp']=dm['timestamp'].apply(timestamp2date)
lv['timestamp']=lv['timestamp'].apply(timestamp2date)
dm=dm.drop(dm[dm.nb_tweets_madrid<200].index) #filtré, muchos datos :c
lv=lv.drop(lv[lv.nb_tweets_liverpool<200].index)
dl=pd.merge(dm,lv,on='timestamp')
dl.plot(kind="Bar",stacked=False,x="timestamp")
#Se habla mas en el minuto 16:39
```

<p>3) En un dataframe, mostrar el número de tweets según el país indicado en el campo
‘location’. ¿Desde cuáles países se enviaron los tweets? (BONUS: mostrar los
resultados en una mapa).</p>

```python
rescountry = es.search(index="tp5-championstweets",body={
  "aggs": {
    "group_by_state": {
      "terms": {
        "field": "location.keyword"
      }
    }
  }
},"query": {"match_all": {}})
countryframe=pd.DataFrame(rescountry['aggregations']['group_by_state']['buckets'])
countryframe=countryframe.rename(columns={'key': 'location', 'doc_count': 'numerotweets'})
countryframe.plot(kind="bar",x="location")



```

<p>4) En un histograma o en un ​ line plot, ​ visualizar dos series de datos:
a) el número de tweets enviado desde Inglaterra, por cada minuto.
b) el número de tweets enviado desde España, por cada minuto.</p>

```python
resinglaterra = es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "1m" \
                                                                                 }}},"query": {"match": {'location':'United Kingdom'}}})
resespaña = es.search(index="tp5-championstweets", body={"aggs" : { "tweets_over_time" : \
                                                            { "date_histogram" : { "field" : \
                                                                                  "timestamp", "interval" : "1m" \
                                                                                 }}},"query": {"match": {'location':'Spain'}}})
esp=pd.DataFrame(resespaña['aggregations']['tweets_over_time']['buckets'])
ing=pd.DataFrame(resinglaterra['aggregations']['tweets_over_time']['buckets'])
esp=esp.rename(columns={'key': 'timestamp', 'doc_count': 'numerotweets_españa'})
ing=ing.rename(columns={'key': 'timestamp', 'doc_count': 'numerotweets_inglaterra'})
esp['timestamp']=esp['timestamp'].apply(timestamp2date)
ing['timestamp']=ing['timestamp'].apply(timestamp2date)
esing=pd.merge(esp,ing,on='timestamp')

esing[:20].plot(kind="bar",x='timestamp')
```
